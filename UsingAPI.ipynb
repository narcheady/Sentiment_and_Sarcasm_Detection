#### Creating a loop for fetching historical tweets

```python
finalTweetTable=pd.DataFrame()
mid=0
bool = True
i = 0
notfoundcount= 0
#for i in range(2):
while bool :
    if i==0:
        tweetList = api.search.tweets(q="coronavirus mental health lang:en -filter:retweets",tweetmode='extended',count=100)
        i += 1
    else:        
        tweetList = api.search.tweets(q="coronavirus mental health lang:en -filter:retweets",tweetmode='extended',count=100,max_id=mid)
    if len(tweetList)>0:
        tweetTable=pd.json_normalize(tweetList['statuses'])
        if 'text' in tweetTable:
            tweetTable = tweetTable[['id','text','created_at']]
            tweetTable['Date'] = pd.to_datetime(tweetTable.created_at)
            mid=tweetTable['id'].min()
            mid=mid-1
            finalTweetTable=pd.concat([finalTweetTable,tweetTable],ignore_index=True)
            if finalTweetTable["Date"].iloc[-1].year < 2020 :
                bool = False
        else:
            notfoundcount += 1
```

```python
accented_string = u'ସପ୍ତାହକରେ ବିଶ୍ବର ପାଖାପାଖି ୫୦ ପ୍ରତିଶତ ସଂକ୍ରମିତ'
```

```python
import unidecode
```

```python
unaccented_string = unidecode.unidecode(accented_string)
unaccented_string
```

```python
import re
import string
```

```python
text = unaccented_string
text = unidecode.unidecode(text)
text = re.sub('https?:\/\/\S+', '', text) # Removing hyperlink
text = re.sub('@[A-Za-z0–9]+', '', text) #Removing @mentions
text = text.translate(str.maketrans('', '', string.punctuation)) #removes punctuation by mapping all punctuations with ''
text = re.sub('#', '', text) # Removing '#' hash tag
text = re.sub('RT[\s]+', '', text) # Removing RT in retweets
text
```
